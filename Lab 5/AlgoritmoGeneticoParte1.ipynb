# Algoritmos Genéticos

## Introducción

Los algoritmos genéticos (AG) son una clase de algoritmos de optimización y búsqueda basados en los principios de la genética y la selección natural de Darwin. Forman parte de los algoritmos evolutivos y se utilizan para resolver problemas complejos donde los métodos tradicionales pueden ser ineficientes o inadecuados. Su capacidad para explorar grandes espacios de búsqueda y encontrar soluciones óptimas o cercanas al óptimo los hace valiosos en diversas aplicaciones.

## ¿Qué es un algoritmo genético?

Un algoritmo genético es un método de optimización inspirado en el proceso de evolución biológica. Emula la forma en que las poblaciones de organismos evolucionan a través de generaciones mediante mecanismos de selección natural, cruce y mutación. El objetivo es encontrar la mejor o las mejores soluciones a un problema específico, representado mediante una función de aptitud que evalúa la calidad de cada solución candidata.

## ¿Cómo funciona un algoritmo genético?

El funcionamiento de un algoritmo genético se basa en los siguientes pasos fundamentales:

1. **Inicialización de la población**: Se crea una población inicial de soluciones candidatas (individuos) de forma aleatoria o utilizando conocimientos previos.

2. **Evaluación**: Cada individuo de la población es evaluado mediante la función de aptitud, que asigna un valor numérico representando su calidad o adaptación al problema.

3. **Selección**: Se eligen los individuos más aptos para reproducirse. Los métodos de selección pueden ser proporcionales a la aptitud, torneos, entre otros.

4. **Cruce (recombinación)**: Los individuos seleccionados (padres) se combinan para producir nuevos individuos (descendientes), intercambiando partes de su información genética. Esto permite explorar nuevas áreas del espacio de búsqueda.

5. **Mutación**: Se aplican cambios aleatorios a los descendientes para mantener la diversidad genética y evitar la convergencia prematura hacia soluciones subóptimas.

6. **Reemplazo**: Los nuevos individuos sustituyen parcial o totalmente a la población actual, formando una nueva generación.

7. **Iteración**: Se repiten los pasos de evaluación, selección, cruce y mutación durante varias generaciones hasta cumplir con un criterio de parada, como alcanzar un número máximo de generaciones o una aptitud satisfactoria.

## Componentes clave de un algoritmo genético

- **Población**: Conjunto de individuos que representan posibles soluciones al problema. El tamaño de la población afecta la diversidad y la convergencia del algoritmo.

- **Representación genética**: Forma en que se codifican las soluciones, comúnmente como cadenas de bits, números reales, permutaciones, entre otros, dependiendo del problema.

- **Función de aptitud**: Mide la calidad de cada individuo en relación con el objetivo del problema. Es esencial para guiar el proceso evolutivo hacia mejores soluciones.

- **Operadores genéticos**:
  - **Selección**: Mecanismo para elegir los individuos que participarán en la reproducción. Favorece a los más aptos, pero permite que individuos menos aptos también tengan oportunidad para mantener la diversidad.
  - **Cruce**: Combina información genética de dos o más padres para producir descendientes. Métodos comunes incluyen cruce de un punto, de dos puntos y uniforme.
  - **Mutación**: Introduce variaciones aleatorias en los individuos. Ayuda a explorar nuevas soluciones y a evitar la estancamiento en óptimos locales.

- **Parámetros del algoritmo**: Incluyen la tasa de cruce, tasa de mutación, tamaño de la población y criterios de parada, que deben ser ajustados cuidadosamente para el problema en cuestión.

## ¿Cómo aprende un algoritmo genético?

El algoritmo genético aprende y mejora las soluciones a través de un proceso iterativo que simula la evolución natural:

- **Selección de los más aptos**: Al favorecer la reproducción de individuos con mayor aptitud, el algoritmo incrementa la probabilidad de que las características beneficiosas se transmitan a las siguientes generaciones.

- **Recombinación de información**: El cruce permite combinar las mejores características de diferentes individuos, explorando nuevas soluciones que podrían ser superiores a las anteriores.

- **Introducción de variabilidad**: La mutación añade diversidad genética, permitiendo al algoritmo escapar de óptimos locales y explorar regiones inexploradas del espacio de búsqueda.

- **Adaptación progresiva**: A lo largo de las generaciones, la población se adapta al problema, mejorando gradualmente la calidad de las soluciones hasta alcanzar o aproximarse al óptimo global.

## Aplicaciones de los algoritmos genéticos

Los algoritmos genéticos se han aplicado exitosamente en diversos campos:

- **Optimización de funciones complejas**: Donde las funciones no son derivables o presentan múltiples óptimos locales.

- **Diseño y ingeniería**: Optimización de estructuras, circuitos electrónicos, diseño aerodinámico, entre otros.

- **Inteligencia artificial y aprendizaje automático**: Entrenamiento de redes neuronales, selección de características y ajuste de hiperparámetros.

- **Problemas combinatorios**: Como el problema del viajante, planificación de rutas y programación de tareas.

- **Bioinformática**: Análisis de secuencias genéticas, predicción de estructuras proteicas y estudios evolutivos.

## Ventajas de los algoritmos genéticos

- **Robustez y flexibilidad**: Pueden aplicarse a una amplia gama de problemas sin requerir información específica del gradiente o derivadas.

- **Búsqueda global**: Su capacidad para explorar múltiples regiones del espacio de búsqueda reduce la probabilidad de quedar atrapados en óptimos locales.

- **Adaptabilidad**: Pueden ajustarse y combinarse con otros métodos para mejorar su rendimiento en problemas específicos.

## Limitaciones de los algoritmos genéticos

- **Alto costo computacional**: Requieren evaluar muchas soluciones en cada generación, lo que puede ser costoso para funciones de aptitud complejas.

- **Parámetros sensibles**: El rendimiento depende de la configuración adecuada de los parámetros, lo cual puede requerir experimentación.

- **No garantizan optimalidad**: Aunque pueden aproximarse al óptimo global, no siempre lo aseguran, especialmente en problemas muy complejos.

> Algoritmo Principal


```
    WHILE Generation < MaxGenerations AND NOT TerminationConditionMet DO
        // Selection
        MatingPool ← SelectIndividuals(Population)
        
        // Crossover (Recombination)
        OffspringPopulation ← []
        FOR i FROM 0 TO (PopulationSize - ElitismCount) STEP 2 DO
            Parent1 ← MatingPool.Select()
            Parent2 ← MatingPool.Select()
            IF Random() < CrossoverRate THEN
                Child1, Child2 ← Crossover(Parent1, Parent2)
            ELSE
                Child1 ← Parent1.Clone()
                Child2 ← Parent2.Clone()
            END IF
            OffspringPopulation.Add(Child1)
            OffspringPopulation.Add(Child2)
        END FOR
        
        // Mutation
        FOR each individual IN OffspringPopulation DO
            IF Random() < MutationRate THEN
                Mutate(individual)
            END IF
        END FOR
        
        // Evaluate Offspring
        FOR each individual IN OffspringPopulation DO
            individual.Fitness ← EvaluateFitness(individual)
        END FOR
        
        // Replacement (Creating New Population)
        Population ← ApplyElitism(Population, OffspringPopulation, ElitismCount)
        
        // Increment generation counter
        Generation ← Generation + 1
    END WHILE

    // Return the best solution found
    BestIndividual ← FindBestIndividual(Population)
    RETURN BestIndividual
END GeneticAlgorithm
```

# Diseño de un Algoritmo Genético para Camuflaje de un Cubo en un Entorno

## Introducción

El objetivo es desarrollar un algoritmo genético que ajuste el color de un cubo para que se camufle con el entorno, específicamente, que su color en escala de grises se acerque al del fondo. Para lograr esto, utilizaremos una población inicial y aplicaremos los principios de los algoritmos genéticos para evolucionar hacia la solución óptima.

La población tendrá un tamaño del 100% (se asume que es un número fijo, por ejemplo, 100 individuos), y en cada generación, solo sobrevivirán el 20% de los individuos cuya representación de color esté más cercana al valor de gris del fondo. Este proceso se repetirá hasta que el cubo alcance un nivel de camuflaje satisfactorio.

## Componentes del Algoritmo Genético

Para resolver este problema, es necesario definir claramente cada uno de los componentes del algoritmo genético:

### 1. Representación Genética (Codificación)

- **Genotipo**: Cada individuo en la población representa un posible color del cubo.
- **Fenotipo**: El color visible del cubo en escala de grises.
- **Codificación**: Utilizaremos un valor numérico que representa el nivel de gris, donde 0 es negro y 255 es blanco.

### 2. Población Inicial

- **Tamaño de la Población**: 100 individuos.
- **Inicialización**: Los valores de gris de los individuos se generan aleatoriamente entre 0 y 255.

### 3. Función de Aptitud (Fitness Function)

- **Objetivo**: Minimizar la diferencia absoluta entre el nivel de gris del individuo y el nivel de gris del fondo.
- **Función de Aptitud**: \( \text{Aptitud}(i) = -|\text{Gris}_{\text{individuo}} - \text{Gris}_{\text{fondo}}| \)
- **Nota**: Al usar el valor negativo de la diferencia, los individuos con menor diferencia tendrán mayor aptitud.

### 4. Selección

- **Criterio de Supervivencia**: Solo el 20% de los individuos con mayor aptitud sobreviven para la siguiente generación.
- **Método de Selección**: Selección elitista, donde se conservan los mejores individuos.

### 5. Operadores Genéticos

#### a. Cruce (Crossover)

- **Tipo de Cruce**: Cruce uniforme.
- **Proceso**: Se seleccionan pares de padres y se combinan sus valores de gris para producir descendientes.

#### b. Mutación

- **Probabilidad de Mutación**: Pequeña, por ejemplo, 1%.
- **Proceso**: Se añade o resta un valor aleatorio pequeño al nivel de gris del individuo.

### 6. Criterio de Parada

- **Condiciones**:
  - Alcanzar un nivel de aptitud satisfactorio (por ejemplo, una diferencia de gris menor a un umbral).
  - Llegar a un número máximo de generaciones.

## Pasos del Algoritmo Genético

1. **Inicialización**: Crear una población inicial de 100 individuos con valores de gris aleatorios entre 0 y 255.

2. **Evaluación de la Aptitud**: Calcular la aptitud de cada individuo utilizando la función de aptitud definida.

3. **Selección de Sobrevivientes**: Retener el 20% de los individuos con mayor aptitud.

4. **Generación de Nueva Población**:
   - **Cruce**: Utilizar los individuos sobrevivientes para generar nuevos individuos mediante cruce.
   - **Mutación**: Aplicar mutación a algunos de los nuevos individuos para mantener la diversidad genética.

5. **Reemplazo**: La nueva población reemplaza a la antigua, manteniendo el tamaño total en 100 individuos.

6. **Verificación del Criterio de Parada**: Si se cumple el criterio de parada, finalizar el algoritmo; de lo contrario, volver al paso 2.

Esta clase se creo con el propóstio de crear objetos que se usarán en el caso del Camaleón.

````python
class Cuadrado(object): 

    def __init__(self, chrom): 
        self.chrom = chrom
        self.fitness = self.cal_fitness()
  
    @classmethod
    def mutated_genes(self):
        gene = rd.randint(0,255)
        return gene 
  
    @classmethod
    def create_gnome(self):
        global obj
        gnome_len = 3
        return np.array([self.mutated_genes() for _ in range(gnome_len)])
  
    def mate(self, par2):
        ''' 
        Perform mating and produce new offspring 
        '''
  
        # chromosome hijo
        child_chromosome = []
        for gp1, gp2 in zip(self.chrom, par2.chrom):     
  
            # prob mate 
            prob = rd.random() 
  
            # if prob is less than 0.45, insert gene 
            # from parent 1  
            if prob < 0.45: 
                child_chromosome.append(gp1) 
  
            # if prob is between 0.45 and 0.90, insert 
            # gene from parent 2 
            elif prob < 0.90: 
                child_chromosome.append(gp2) 
  
            # otherwise insert random gene(mutate),  
            # for maintaining diversity 
            else: 
                child_chromosome.append(self.mutated_genes()) 
  
        # create new Individual(offspring) using  
        # generated chromosome for offspring 
        return Cuadrado(np.array(child_chromosome)) 
  
    def cal_fitness(self): 
        # Calculo fitness
        global obj
        fitness = 0
        dif = self.chrom - obj
        fitness = (dif[0]**2+dif[1]**2+dif[2]**2)
        return fitness 
````

Aquellas variables que están fuera de la función main() son variables globales que se usarán a lo largo de la ejecución. De estas, la que cabe resaltar es la variable obj es el objetivo al que los individuos deben llegar. Por otra parte, max_gen es el número de generaciones máximo que se ejecutarán. Finlamente, la variable history se usa para guardar las generaciones para posteriormente gráficarlas. 

El método main() se ejecuta para iterar las generaciones, buscando aquellos individuos que son los más aptos. Adicionalmente, en este método, cada 40 generaciones, se va cambiando el color del fondo, esto para que los individuos se adapten a nuevas condiciones. Por último, esta se encarga de hacer los cruces necesarios para pasar de una generación a otra.

````python
# Target string to be generated
global obj
obj = np.array([rd.randint(0,255),rd.randint(0,255),rd.randint(0,255)])
max_gen = 400
history = []
  
# Driver code 
def main(): 
    POPULATION_SIZE = 100
  
    #current generation 
    generation = 1
  
    found = False
    population = [] 
  
    # create initial population 
    for _ in range(POPULATION_SIZE):
        gnome = Cuadrado.create_gnome() 
        population.append(Cuadrado(gnome))
  
    while not found and generation != max_gen:

        if generation%40 == 0:
            global obj
            obj = np.array([rd.randint(0,255),rd.randint(0,255),rd.randint(0,255)])
            for sqr in population:
                sqr.fitness = sqr.cal_fitness()
  
        # sort the population in increasing order of fitness score 
        population = sorted(population, key = lambda x:x.fitness) 

        #for i in population:
        #    print(i.fitness)
  
        # if the individual having lowest fitness score ie.  
        # 0 then we know that we have reached to the target 
        # and break the loop 
        if population[0].fitness <= -1: 
            found = True
            break
  
        # Otherwise generate new offsprings for new generation 
        new_generation = [] 
  
        # Perform Elitism, that mean 10% of fittest population 
        # goes to the next generation 
        s = int((20*POPULATION_SIZE)/100) 
        new_generation.extend(population[:s]) 
  
        # From 50% of fittest population, Individuals  
        # will mate to produce offspring 
        s = int((80*POPULATION_SIZE)/100) 
        for _ in range(s): 
            parent1 = rd.choice(population[:50]) 
            parent2 = rd.choice(population[:50]) 
            child = parent1.mate(parent2) 
            new_generation.append(child) 
  
        population = new_generation 
  
        print("Generation: {}\tString: {}\tFitness: {}".\
              format(generation, 
              "".join(str(population[0].chrom)),
              population[0].fitness))
  
        history.append(new_generation)

        generation += 1

    print("Generation: {}\tString: {}\tFitness: {}".\
          format(generation, 
          "".join(str(population[0].chrom)), 
          population[0].fitness))
  
````

Esta función se encarga de gráfica las generaciones.

````python
# Visualization
def plot_evolution(history):
    fig, axes = plt.subplots(len(history), 1, figsize=(10, len(history) * 0.5))
    for gen_idx, generation in enumerate(history):
        colors = [tuple(ind/255 for ind in sqr.chrom) for sqr in generation]
        axes[gen_idx].imshow([colors], aspect="auto")
        axes[gen_idx].axis("off")
        axes[gen_idx].set_title(f"Generation {gen_idx + 1}")

    plt.tight_layout()
    plt.show()

plot_evolution(history)
````
